```python
import gmaps
import pandas as pd

# Google developer API key
from config import gkey

# Configure gmaps
gmaps.configure(api_key=gkey)
```

```python
# Create aiport dataframe
data = pd.read_csv('final_10_neighborhoods.csv')
data.dropna(inplace = True) 
#data.head()
#has duplicates so separate csv's are imported for analyze AirBnb and Regular properties separately
```

```python
#import airbnb data - airbnb data has mostly 20 properties per neighborhood, but 4 neighborhoods have more
#we we will randomly select 20 properties from each of these to ensure the heat maps display controlled and accurate data
air_data = pd.read_csv('air_neighborhoods.csv')
data.dropna(inplace = True) 
#loc to HOUSE - not used as we opted to show all types of properties
air_house = air_data.loc[
    (air_data["AirBnb Type"] == "House")]
air_house.head()
```

```python
#South LA (106 to remove), Westwood (4 to remove), Downtown(33 to remove), Hollywood (2 to remove)
#south los angeles
air_south_la = air_data.loc[
    (air_data["neighborhood name"] == "South Los Angeles")
]
la_properties = air_south_la.sample(20)

#westwood
air_westwood = air_data.loc[
    (air_data["neighborhood name"] == "Westwood")
]
westwood_properties = air_westwood.sample(20)

#downtown
air_downtown = air_data.loc[
    (air_data["neighborhood name"] == "Downtown")
]
downtown_properties = air_downtown.sample(20)

#hollywood
air_hollywood = air_data.loc[
    (air_data["neighborhood name"] == "Hollywood")
]
hollywood_properties = air_hollywood.sample(20)
```

```python
pd.set_option("display.max_rows", 40, "display.max_columns", None)
```

```python
#replace air data for those three neighborhoods with random data
# air_data.loc[
#     (air_data["neighborhood name"] == "South Los Angeles")]

#remove the three neighborhoods
air_data2 = air_data
air_data2 = air_data2.drop(air_data2[air_data2["neighborhood name"] == "South Los Angeles"].index)
air_data2 = air_data2.drop(air_data2[air_data2["neighborhood name"] == "Westwood"].index)
air_data2 = air_data2.drop(air_data2[air_data2["neighborhood name"] == "Downtown"].index)
air_data2 = air_data2.drop(air_data2[air_data2["neighborhood name"] == "Hollywood"].index)

#add the random neighborhoods
air_data2 = air_data2.append(la_properties)
air_data2 = air_data2.append(westwood_properties)
air_data2 = air_data2.append(downtown_properties)
air_data2 = air_data2.append(hollywood_properties)
air_data2.head()
```

```python
#import regular data - regular data has 20 properties per neighborhood
reg_data = pd.read_csv('reg_neighborhoods.csv')
data.dropna(inplace = True) 
reg_data.head()
#loc to HOUSE - not used as we opted to show all types of properties
reg_house = reg_data.loc[
    (reg_data["Regular Type"] == "single_home")|
    (reg_data["Regular Type"] == "townhouse")]
reg_house.head()
```

```python
#group data to show one marker per neighborhood and average Airbnb property price
grouped_air_data = air_data2[['neighborhood name', 'AirBnb Lat', 'AirBnb Long', 'AirBnb Price']].groupby('neighborhood name').mean()
grouped_air_data
```

```python
#create hover text variable
hover_air = grouped_air_data.index + "\n" + "Average Airbnb monthly income is $" + grouped_air_data['AirBnb Price'].apply(str)
hover_air
```

```python
# Store latitude and longitude in locations
air_neighborhoods = air_data2[["AirBnb Lat", "AirBnb Long"]]
reg_neighborhoods = reg_data[["Regular Lat", "Regular Long"]]

# Store latitude and longitude in locations - HOUSE
air_neighborhoods_house = air_house[["AirBnb Lat", "AirBnb Long"]]
reg_neighborhoods_house = reg_house[["Regular Lat", "Regular Long"]]

# Fill NaN values and convert to float
air_price = air_data2["AirBnb Price"].astype(float)/10000
reg_price = reg_data["Regular Price"].astype(float)/10000

#price variances for just house
air_house_price = air_house["AirBnb Price"].astype(float)/1000
reg_house_price = reg_house["Regular Price"].astype(float)/1000
#181 observations for reg house data and price
#102 observations for air house data and price
```

```python
# Plot Heatmap AIRBNB - ALL PROPERTIES
air_fig = gmaps.figure(center=[34.00510025,-118.4349976], zoom_level=11)
#
# Create heat layer AIRBNB
heat_layer = gmaps.heatmap_layer(air_neighborhoods, 
                                weights=air_price, 
                                 dissipating=True,
                                 point_radius=30)

marker_layer_air = gmaps.marker_layer(grouped_air_data[['AirBnb Lat', 'AirBnb Long']],
                                hover_text= hover_air,
                                 label=grouped_air_data.index,
                                info_box_content=None,
                                display_info_box=None)




# Add layer
air_fig.add_layer(heat_layer)
air_fig.add_layer(marker_layer_air)

# Display figure
air_fig
```

```python
#group data to show one marker per neighborhood and average regular property price
grouped_reg_data = reg_data[['neighborhood name', 'Regular Lat', 'Regular Long', 'Regular Price']].groupby('neighborhood name').mean()
grouped_reg_data
```

```python
#create hover text variable
hover_reg = grouped_reg_data.index + "\n" + "Average regular property price per month $" + grouped_reg_data['Regular Price'].apply(str)
hover_reg
```
